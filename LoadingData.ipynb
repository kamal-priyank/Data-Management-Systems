# Create git issue table

```python
# print data column and column type

issue_type = {}

for col in gitissue.columns:
    issue_type[col] = type(gitissue.loc[0,col])

issue_type
```

```python
# navigate to create table columns

create_gitissue_command = ("""
        DROP TABLE IF EXISTS GITISSUE;
        CREATE TABLE GITISSUE(
            id                  INT,
            title               TEXT,
            state               VARCHAR(6),
            body                TEXT,
            _user               VARCHAR(100),
            user_id             VARCHAR(100),
            repository          FLOAT(32),
            created_at          TIMESTAMP WITHOUT TIME ZONE,
            updated_at          TIMESTAMP WITHOUT TIME ZONE,
            closed_at           TIMESTAMP WITHOUT TIME ZONE,
            assignees           VARCHAR(100),
            labels              TEXT,
            reactions           VARCHAR(65000),
            n_comments          SMALLINT,
            closed_by           VARCHAR(100),
            comment_id          INT,
            comment_created_at  TIMESTAMP WITHOUT TIME ZONE,
            comment_updated_at  TIMESTAMP WITHOUT TIME ZONE,
            comment_user_id     VARCHAR(100),
            comment_user        VARCHAR(100),
            comment_text        TEXT,
            project             VARCHAR(20)   
        )""")

cursor.execute(create_gitissue_command)
    
connection.commit()
```

```python
# define function import data

def insert_into_gitissue(cursor, id, title, state, body, user, user_id, repository, created_at,
       updated_at, closed_at, assignees, labels, reactions,
       n_comments, closed_by, comment_id, comment_created_at,
       comment_updated_at, comment_user_id, comment_user, comment_text,
       project):
       
       insert_into_comment = ("""
       INSERT INTO gitissue (
       id, title, state, body, _user, user_id, repository, created_at,
       updated_at, closed_at, assignees, labels, reactions,
       n_comments, closed_by, comment_id, comment_created_at,
       comment_updated_at, comment_user_id, comment_user, comment_text,
       project) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s);
       """)
       
       row_to_insert = (
       id, title, state, body, user, user_id, repository, created_at,
       updated_at, closed_at, assignees, labels, reactions,
       n_comments, closed_by, comment_id, comment_created_at,
       comment_updated_at, comment_user_id, comment_user, comment_text,
       project)
       
       cursor.execute(insert_into_comment,row_to_insert)
       
       connection.commit()
```

```python
# insert data to gitissue table

for i, row in gitissue.iterrows():
    insert_into_gitissue(cursor, row["id"],  
                        row['title'],
                        row['state'],
                        row['body'],
                        row['user'], 
                        row['user_id'], 
                        row['repository'], 
                        row['created_at'], 
                        row['updated_at'],
                        row['closed_at'], 
                        row['assignees'], 
                        row['labels'], 
                        row['reactions'], 
                        row['n_comments'],
                        row['closed_by'], 
                        row['comment_id'], 
                        row['comment_created_at'], 
                        row['comment_updated_at'],
                        row['comment_user_id'], 
                        row['comment_user'], 
                        row['comment_text'], 
                        row['project'])
```

# Creat git data table

```python
# for loop insert data into gitdata table

data_type = {}

for col in gitdata.columns:
    data_type[col] = type(gitdata.loc[0,col])

data_type
```

```python
# naviagte to create table

create_gitdata_command = ("""
        DROP TABLE IF EXISTS GITDATA;
        CREATE TABLE GITDATA(
        id INT,
        hash TEXT, 
        msg TEXT, 
        author_name TEXT, 
        committer_name TEXT, 
        author_date TIMESTAMP WITHOUT TIME ZONE,
        author_timezone INT, 
        committer_date TIMESTAMP WITHOUT TIME ZONE, 
        committer_timezone INT, branches TEXT,
        in_main_branch BOOL, 
        merge BOOL, 
        parents TEXT, 
        project_name TEXT, 
        deletions INT,
        insertions INT, 
        lines INT, files INT, 
        old_path TEXT, 
        new_path TEXT, 
        filename TEXT,
        change_type TEXT, 
        diff TEXT, 
        diff_parsed TEXT, 
        deleted_lines INT, 
        source_code TEXT,
        source_code_before TEXT, 
        nloc FLOAT(32), 
        complexity FLOAT(32),
        token_count FLOAT(32)   
        );""")
cursor.execute(create_gitdata_command)

connection.commit()
```

```python
# define function to import data

def insert_into_gitdata(cursor, id, hash, msg, author_name, committer_name, author_date,
                        author_timezone, committer_date, committer_timezone, branches,
                        in_main_branch, merge, parents, project_name, deletions,
                        insertions, lines, files, old_path, new_path, filename,
                        change_type, diff, diff_parsed, deleted_lines, source_code,
                        source_code_before, nloc, complexity, token_count):
    insert_into_comment = ("""
              INSERT INTO gitdata (
              id, hash  , msg  , author_name  , committer_name  , author_date  ,
              author_timezone  , committer_date  , committer_timezone  , branches  ,
              in_main_branch  , merge  , parents  , project_name  , deletions  ,
              insertions  , lines  , files  , old_path  , new_path  , filename  ,
              change_type  , diff  , diff_parsed  , deleted_lines  , source_code  ,
              source_code_before  , nloc  , complexity  , token_count ) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s);
              """)
    row_to_insert = (
        id, hash, msg, author_name, committer_name, author_date,
        author_timezone, committer_date, committer_timezone, branches,
        in_main_branch, merge, parents, project_name, deletions,
        insertions, lines, files, old_path, new_path, filename,
        change_type, diff, diff_parsed, deleted_lines, source_code,
        source_code_before, nloc, complexity, token_count)
    cursor.execute(insert_into_comment, row_to_insert)
    connection.commit()
```

```python
# insert data to gitdata table

for i, row in gitdata.iterrows():
    insert_into_gitdata(cursor, row['id'],row['hash'],
                        row['msg'],
                        row['author_name'],
                        row['committer_name'],
                        row['author_date'],
                        row['author_timezone'],
                        row['committer_date'],
                        row['committer_timezone'],
                        row['branches'],
                        row['in_main_branch'],
                        row['merge'],
                        row['parents'],
                        row['project_name'],
                        row['deletions'],
                        row['insertions'],
                        row['lines'],
                        row['files'],
                        row['old_path'],
                        row['new_path'],
                        row['filename'],
                        row['change_type'],
                        row['diff'],
                        row['diff_parsed'],
                        row['deleted_lines'],
                        row['source_code'],
                        row['source_code_before'],
                        row['nloc'],
                        row['complexity'],
                        row['token_count'])
```
